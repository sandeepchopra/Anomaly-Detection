```python
import numpy as np
import pandas as np
import matplotlib.pyplot as plt
from sklearn.datasets.samples_generator import make_blobs
from sklearn.cluster import KMeans
%matplotlib inline
```

```python
X, y = make_blobs(n_samples=500, centers=5, n_features=2,random_state=1)
print y
plt.scatter(X[:,0],X[:,1])
```

```python
clusters=[]
within_sum_of_squares=[]
for i in range(1,20):
 model=KMeans(n_clusters=i)
 model.fit_predict(X)
 # print model.score
 clusters.append(model.n_clusters)
 # print model.cluster_centers_
 within_sum_of_squares.append(model.inertia_)
plt.plot(clusters,within_sum_of_squares)
```
